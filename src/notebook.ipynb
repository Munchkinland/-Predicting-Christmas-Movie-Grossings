ðŸ–¥Data Preprocessing 

```python
# Dropping rows where the target variable (gross) is missing
data_clean = merged_df_final.dropna(subset=['gross'])
```

```python
# Handling missing values in other columns
# For numerical columns, we'll impute with the median
# For categorical columns, we'll impute with the most frequent value
numerical_cols = data_clean.select_dtypes(include=['float64', 'int64']).columns
categorical_cols = data_clean.select_dtypes(include=['object']).columns
```

```python
# Defining the preprocessing for numerical and categorical data
numerical_transformer = SimpleImputer(strategy='median')
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])
```

```python
# Bundle preprocessing for numerical and categorical data
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_cols),
        ('cat', categorical_transformer, categorical_cols)
    ])
```

Feature Engineering ðŸ‘¨â€ðŸ’»

```python
# Creating a 'decade' feature
data_clean['decade'] = (data_clean['release_year_x'] // 10 * 10).astype(int)
```

```python
# Update categorical_cols to include the new 'decade' feature
categorical_cols = categorical_cols.tolist() + ['decade']
```

```python
# Defining our target variable
y = data_clean['gross']
```

```python
# Define features
X = data_clean[numerical_cols.tolist() + categorical_cols]
```

```python
# Splitting data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Model
# We'll use a RandomForestRegressor as a starting point
model = Pipeline(steps=[('preprocessor', preprocessor),
                        ('model', RandomForestRegressor(n_estimators=100, random_state=0))])
```

```python
# Training the model
model.fit(X_train, y_train)
```

```python
#Save model
file_name = "random_forest_ne100_rs42.joblib"

# Saving model in a file
joblib.dump(model, file_name)
```

```python
# Predicting on test data
y_pred = model.predict(X_test)
```

```python
# Evaluation metrics
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
print("mse", mse, "r2", r2)
```
